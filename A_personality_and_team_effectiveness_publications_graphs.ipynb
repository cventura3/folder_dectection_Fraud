Below is a table for the pvalue of this df

```python
#since we have the correlation lets get the p.value

from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues

calculate_pvalues(bigfive_personality)
#below are the p values
```

See below a table with the pearson correlation using * to show sig at an alpha level of =0.05 and ** to show sig at an alpha level of =0.001

Also 
Pandas dataframe.mask() function return an object of same shape as self and whose corresponding entries are from self where cond is False and otherwise are from other object. The other object could be a scalar, series, dataframe or could be a callable. The mask method is an application of the if-then idiom. For each element in the calling DataFrame, if cond is False the element is used; otherwise the corresponding element from the DataFrame other is used
taken from :https://www.geeksforgeeks.org/python-pandas-dataframe-mask/

```python
#use the pearson r corr to see significance at .05 and .01

rho = bigfive_personality.corr()
rho = rho.round(3)
pval = calculate_pvalues(bigfive_personality) # toto_tico's answer
# create three masks
r1 = rho.applymap(lambda x: '{}'.format(x)) # I do not want the significant pvalue of .1
r2 = rho.applymap(lambda x: '{}*'.format(x))
r3 = rho.applymap(lambda x: '{}**'.format(x))
print(r1.head())
# apply them where appropriate
rho = rho.mask(pval<=0.1,r1)
rho = rho.mask(pval<=0.05,r2)
rho = rho.mask(pval<=0.01,r3)
rho = rho.mask(pval== 0, r1)

rho
# note I prefer readability over the conciseness of code, 
# instead of six lines it could have been a single liner like this:
# [rho.mask(pval<=p,rho.applymap(lambda x: '{}*'.format(x)),inplace=True) for p in [.1,.05,.01]]
#below is the table with the r correlation taking the pvalues
```
