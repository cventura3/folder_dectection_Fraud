```python
#Lets add some color to each boxplot

boxplot= group_effectiveness.boxplot(patch_artist=True, return_type='dict')

for b in boxplot['boxes']:

    b.set(facecolor='gray')


# add a black line to the medians in each boxplot
for median in boxplot['medians']:

    median.set(color='black', linewidth=2)


plt.show()
```

```python
#Lets add some color to each boxplot
#change outliers with sym = '+'
boxplot= group_effectiveness.boxplot(patch_artist=True, return_type='dict', sym = '+')

for b in boxplot['boxes']:

    b.set(facecolor='gray')


# add a black line to the medians in each boxplot
for median in boxplot['medians']:

    median.set(color='black', linewidth=2)


plt.show()

#bp = boxplot.boxplot(data, notch=0, sym='+', vert=1, whis=1.5)
```

```python
boxplot= group_effectiveness.boxplot(patch_artist=True, return_type='dict', sym = '+')

# add a black line to the medians in each boxplot
for median in boxplot['medians']:

    median.set(color='black', linewidth=2)
    
# you can do different modifications to the boxplot

plt.setp(boxplot['boxes'], color='blue')             # change color inside of the boxes
plt.setp(boxplot['whiskers'], color='red')           #change line color 
plt.setp(boxplot['fliers'], color='green', marker='+')
plt.show()
```

```python
boxplot= group_effectiveness.boxplot(patch_artist=True, return_type='dict', sym = '+') 


# you can change the outline for the boxplot
for box in boxplot ['boxes']:
    # change outline color
    box.set(color='red', linewidth=2)
    # change fill color
    box.set(facecolor = 'green' )
    # change hatch      # you can also use these for hatch ('-', '+', 'x', '\\', '*', 'o', 'O', '.')
    box.set(hatch = '*')

plt.show()
#amazing!
```

```python
#lets try another example changing each boxplot color!
bp= group_effectiveness.boxplot(patch_artist=True, return_type='dict', sym = '+')

# fill with colors
colors = ['pink', 'lightblue', 'lightgreen','grey','pink', 'red','red', 'red']

for x, col in zip(bp['boxes'], colors):
    x.set_facecolor(col)

plt.show ()

```
